## USER GUIDE: 

-> Listed below are the steps to be followed to sell the products from the inventory:

> First the owner needs to open the inventory for the day.

> When you run the following code, at first it would ask a you to enter a number to open inventory.

> PLEASE NOTE THAT YOU CAN ENTER ANY NUMBER TO OPEN THE INVENTORY EXCEPT -321.

> Once the inventory is open for the customers you just need to sit back and let the system do all the work for you.

> If you wish to close inventory for the day enter the product id as -321. 

> Your SALES.json will be updated with every customer whereas the record.json will be updated after you close the inventory. Also you can manually update it by running the last cell.


-> Listed below are a few steps for the customers.

> When you visit the system, you will be shown a WELCOME message followed by the list of products present in the inventory.

>To make a purchase a customer needs to follow GENERAL INSTRUCTIONS mentioned in the following code.

> Once the customer has added the stuff he/she desire to buy he/she need to enter the PRODUCT ID as -1 to proceed for checkout.

> After the customer has checkout out he/she will see the items that he/she have added to the cart and then the final bill will be generated.

> The customer then can cross check the items and can download the BILL.json file from the FILES tab present to the LEFT SIDE of the EDITOR window.

> Once the customer has downloaded the bill he/she can press ENTER and then the screen will be cleared for the next customer in 20 seconds.

```python
#    Importing the json library to use json.dumps() 
#    and json.loads() function in alter part of the code
 
import json
import time
import pprint
from google.colab import output
 
  
# Opening record.json file
 
with open('record.json') as json_file:
    data = json.load(json_file)
 
# To open the Inventory for customers: press -123 
 
# Initializing transaction ID to zero at beginning of the sales
 
transaction_ID=0
 
# Serializing json 
json_object = json.dumps({}, indent = 4)
  
# Creating SALES.json to save the sales for the day
with open("SALES.json", "w") as outfile:
    outfile.write(json_object)
id = int(input("ENTER ANY NUMBER TO OPEN THE INVENTORY: "))
 
# Initializing the dict to store the sales for the day

 
print('Opening the Inventory for customers\n\n')
#Enter -321 as ID to Close the inventory
output.clear()
while id != '-321':
  print("\n\n\n******WELCOME TO THE CADBUARY INVENTORY******\n\n\n")
  sales = {}
  # General Instructions TO The Customer:
  print('***General Instructions***\n\n')
  print('-> Enter the ID of the product that you want to add to the cart:\n')
  print('-> If you wish to remove any of the previously added item from the cart, please Enter 0 as ID:\n')
  print('-> If you wish to proceed to Checkout, please enter -1 as ID:\n')
  print('-> Dont forget to download your BILL!\n')
 
# showing the Inventory present to the customer
  print("The following items are prosent in the Inventory: \n")
  
  pprint.pprint(data)
 
# initializing the bill and cart for the customer
  bill=0
  cart={}
  
# Lets Sell!!
  while True:
    id=input("\n\nPlease Enter the id of product that you want to buy: ")
    if id=='-321':                                                      # if you wish to close the inventory for the day
      break
 
    if id=='-1':                                                        # if customer wish to proceed to checkout  
      print("THANKS FOR SHOPING")
      break
    
    if id=='0':                                                         # if customer want to remove the previously added item from the cart
      id=input("Please Enter the ID of Product that you want to remove from cart: ")
      if id in cart:
        data[id]['Quantity Available']=data[id]['Quantity Available'] + cart[id]['Quantity added to Cart']  # restoring the items removed from the cart to the inventory
        bill=bill-cart[id]['Quantity added to Cart']*data[id]['Price']  # reducing the bill for the items removed    
        print(cart[id]['Name'],"is removed from your cart!\n")
        del cart[id]                                                    # deleting the item from the cart
        continue
      else:
        print('Your cart does not contain the item selected to return!\n') # if the customer wish to return the item which is not added to cart in first place
      continue
    
    if id not in data:                                            
      print('Invalid ID! Please Enter ID again or -1 to Exit\n')          # if ID is not entered correctly
      print("Below are the detail of Products: \n")
      pprint.pprint(data)                                          # Showing the ID and other details to customer
      continue
    print('You have selected',data[id]['Name'])                         # Displaying the Product selected by the customer
    num=int(input("Please Enter the Quantity the you wish to buy: "))   # asking for the quantity of the product they wish to buy
 
    if num<=data[id]['Quantity Available']:                                 # to check if we have more or equal inventory so as to fulfil the need of the customer
      data[id]['Quantity Available']=data[id]['Quantity Available'] - num   # inventory update
      bill=bill+(num*data[id]['Price'])                                     # updating the bill of the customer  
    
      if id not in cart:                                                    # if the customer enters the item for the first time into the cart then creating a record of the tem into the cart
        cart[id]=dict(data[id])
        cart[id]["Quantity added to Cart"]=num
        del cart[id]["Quantity Available"]                                  # the customer need not to know know how many items left after his purchase so that key value pair is removed.
      else:
        cart[id]["Quantity added to Cart"]=cart[id]["Quantity added to Cart"]+num   # if the cart already contain the product then updating the number of items added to cart
      print(num,data[id]['Name'],'added to your cart! ')
    
    else:
      print("Sorry Currently We have only",data[id]['Quantity Available'],' available in our Inventory.') # if inventory has less no of items than requested
 
  if id=='-321': 
    print('Closing Inventory for the day!!!')                                                     # to close the inventory
    break
  print("Your Cart contains items as follows:")                             # show the customer what does their cart consist of
 
  # Print the items present in the cart
 
  print('\n\n')
  print(json.dumps(cart,indent=4))
  print('\n\n')
  
  print("Your Final Bill is: ",bill)                                        # final bill of the customer
 
 
  # Generating bill for the customer
  for id in cart:
    cart[id]["Total Of Product"]=cart[id]['Quantity added to Cart']*cart[id]['Price']
 
  cart['Grand Total']=bill                                                  # adding the final bill to the cart dict so as to mention in the BILL.json file
    
  # Creating the BILL.json 
  with open("BILL.json", "w") as write_file:
      json.dump(cart, write_file, indent=4)
 
 
  transaction_ID= transaction_ID+1                      
  sales[transaction_ID]=dict(cart)                                    # adding transaction record to sales dict
  
  print('\n\n')
  print(json.dumps(cart,indent=4))
  print('\n\n')  
  with open("SALES.json", "r+") as file:                              # updating the SALES.json file
    js = json.load(file)
    js.update(sales)
    file.seek(0)
    json.dump(js, file, indent=4)
 
 
  
  print('Press ENTER and the screen shall be cleared in 20 seconds!!')
  input('')
  print('Thanks for Shopping\nPlease Visit Again')
  time.sleep(20)                            # to let the customer leave and come in next customer
 
  output.clear()
 
# updating record.json file
 
with open('record.json', 'w') as fp:
    json.dump(data, fp,  indent=4)
```
